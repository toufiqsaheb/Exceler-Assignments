```python
import spacy

def pos_tagging(sentence):
    """
    Performs part-of-speech tagging on the given sentence using SpaCy.
    
    Parameters:
    sentence (str): The input sentence to analyze.
    
    Returns:
    list: A list of tuples where each tuple contains a token and its POS tag.
    """
    # Load SpaCy's English language model
    nlp = spacy.load("en_core_web_sm")
    
    # Process the sentence
    doc = nlp(sentence)
    
    # Extract tokens and their POS tags
    pos_tags = [(token.text, token.pos_) for token in doc]
    
    return pos_tags

# Test the function with the given sentence
sentence = 'NLP is amazing and fun to learn.'
tags = pos_tagging(sentence)
for token, pos in tags:
    print(f"{token}: {pos}")
```
